```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
years=[ i for i in range(2007,2020)]
citation=[4,22,47,36,52,41,45,43,32,35,41,43,49]
p=pd.DataFrame(data=citation,index=years)
sns.set(style="whitegrid")
ax=sns.lineplot(years,citation,label="Citas")
ax.set(xlabel='Anos', ylabel='Citas')
plt.show()
```
